## Paths

A **path** is a sequence of edges to get from one node to another

````python
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("F", "A")
g.edge("A", "E")
g
# B,C,D,F is a path from B to F
# B,A,F is also a path from B to F
# FUTURE QUESTION: what is the shortest path?
# there is no path from A to E
````

````python
# "connected" means that given any two nodes, there's a path from the first to the second
# "cycle" which is a path from a node back to itself, with no repeating edges
````

## Directed Graphs, DAGs

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("F", "A")
g.edge("A", "E")
g
# "strongly connected" (same definition as "connected") -- path from any start to any end
# "weakly connected" -- it would be connected if we ignored the direction of the edges
# "parent" -- means the source, or outgoing side of an edge (all of them except E)
# "child" -- means destination, or incoming side of an edge (all of them are children)
````

````python
# DAG:
# 1. Directed
# 2. Acyclic -- means no cycles
# 3. Graph
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("A", "F")
g.edge("A", "E")
g
````
